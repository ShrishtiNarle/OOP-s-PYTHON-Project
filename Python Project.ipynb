```python
class FulltimeEmp(Employee):
    def __init__(self,ename,eadd,mk,md,yofm,mil,esal):
        super().__init__(ename,eadd,mk,md,yofm,mil)
        self.__emp_sal=esal
        
    def get_salary(self):
        return self.__emp_sal
    def set_salary(self,esal):
        self.__emp_sal=esal
        
    def compute(self):
        sal= self.get_salary()
        if sal <= 45000:
            tax= (sal * 18)/100
        elif sal> 45000 and sal<= 82000:
            tax = (45000 * 18)/100 + ((sal -45000)* 28)/100
        elif sal>82000:
            tax = ((45000 * 18)/100 + ((82000 -45000)* 28)/100 + ((sal-82000)* 33)/100)
        else:
            print(" invalid Salary!!")
        return (sal-tax)
        
        
    def __str__(self):
        ftempdata= super().__str__()
        ftempdata+= "Salary of Full time employee: " + self.__emp_sal
        return (ftempdata)
```

```python
class HourEmp(Employee):
    def __init__(self,ename,eadd,mk,md,yofm,mil,hwrk, hrate):
        super().__init__(ename,eadd,mk,md,yofm,mil)
        self.__hrs_worked=hwrk 
        self.__hrs_rate=hrate
    def get_hours_wrkd(self):
        return self.__hrs_worked
    
    def get_hour_rate(self):
        return self.__hrs_rate
    
    def set_hours_wrkd(self, hwrk):
        self.__hrs_worked=hwrk
        
    def set_hour_rate(self, hwrk):
        self.__hrs_rate=hrate
        
    def compute(self):
        hw= self.get_hours_wrkd()
        hr= self.get_hour_rate()
        if hw <= 40:
            return (hw * hr)
        elif hw > 40:
            return ((40 * hr)+ (hw - 40)* hr *1.8)
            
        else:
            return("Invalid hours work")
        
    def __str__(self):
        hremp= super().__str__()
        hremp+= "total number of hours worked: " + self.__hrs_worked + "per hour rate: " + self.__hrs_rate
        return (hremp)
```

```python
class Consultant(Employee):
    def __init__(self,ename,eadd,mk,md,yofm,mil,hwrk,pjtyp):
        super().__init__(ename,eadd,mk,md,yofm,mil)
        self.__hours_wrkd=hwrk
        self.__project_type=pjtyp
    
    def get_conhrs_wrkd(self):
        return self.__hours_wrkd
    
    def get_project_type(self):
        return self.__project_type
    
    def set_conhrs_wrkd(self,hwrk):
        self.__hours_wrkd=hwrk
        
    def set_project_type(self,pjtyp):
        self.__project_type=pjtyp
        
    def compute(self):
        hw= self.get_conhrs_wrkd()
        pj= self.get_project_type()
        
        if pj==1:
            return(55 * hw)
        elif pj==2:
            return(70 * hw)
        elif pj==3:
            return(85 * hw)
        else:
            return("Invalid Project Type")
        
        
    def __str__(self):
        consultdata = super().__str__()
        consultdata+= "total hours worked: " + self.__hours_wrkd() + "Project Type: " + self.__project_type
        return (consultdata)
```
